**Direct Invers Mehode**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([2,5,1,7,9,3,6,10])



# bentuk matriks desain
A = np.vstack([x, np.ones(len(x))]).T

# ubah y menjadi kolom
y = y[:, np.newaxis]

# pseudoinverse
pinv = np.linalg.pinv(A)

# koefisien alpha = [m, c]
alpha = np.dot(pinv, y)

# plot data asli
plt.scatter(x, y, label='Data')

# plot garis regresi
plt.plot(x, alpha[0]*x + alpha[1], label='garis')

plt.legend()
plt.show()
```

Polynomial Regression

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([2,5,1,7,9,3,6,10])

for i in range(1, 7):
  y_est = np.polyfit(x,y,i)

  plt.subplot(2,3,i)
  plt.plot(x,y, 'ro')
  plt.plot(x, np.polyval(y_est, x))
  plt.title(f"No.{i}")
  plt.tight_layout()

plt.show()

```

```python
from scipy.interpolate import CubicSpline

x = np.array([1,2,3,4,5,6,7,8])
y = np.array([2,5,1,7,9,3,6,10])


plt.plot(x,y,'ro')

f = CubicSpline(x,y)

x_new = np.linspace(1, 8 , 10)
y_new = f(x_new)

print(x_new)

plt.plot(x_new, y_new)

```
